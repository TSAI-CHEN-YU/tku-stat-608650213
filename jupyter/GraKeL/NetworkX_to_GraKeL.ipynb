
# Example：transforming NetworkX graphs to GraKeL graphs.



```python
from __future__ import print_function
print(__doc__)

import numpy as np
import networkx as nx

from grakel.utils import graph_from_networkx
```


## 將 NetworkX圖列表 轉換為 GraKeL圖列表
networkx.Functions : https://networkx.org/documentation/stable/reference/functions.html  

grakel.graph_from_networkx : https://ysig.github.io/GraKeL/0.1a8/generated/grakel.graph_from_networkx.html


### node-labeled graphs(GraKeL)


```python
# Creates a list of two node-labeled graphs
G1 = nx.Graph()
G1.add_nodes_from([0,1,2])
G1.add_edges_from([(0,1), (1,2)])
nx.set_node_attributes(G1, {0:'a', 1:'b', 2:'a'}, 'label')

G2 = nx.Graph()
G2.add_nodes_from([0,1,2])
G2.add_edges_from([(0,1), (0,2), (1,2)])
nx.set_node_attributes(G2, {0:'a', 1:'b', 2:'c'}, 'label')

G_nx = [G1, G2]

# 將 NetworkX圖列表 轉換為 GraKeL圖列表
G_1 = graph_from_networkx(G_nx, node_labels_tag='label')
print("1 - Node-labeled graphs transformed")
```

```python
list(G_1)
```

```python
G=G2
# node-attributed graphs 繪圖
plt.figure(dpi=80,figsize=(6, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos)

#Generate node_labels manually
node_labels = {}
for node in G.nodes:
    node_labels[node] = G.nodes[node] #G.nodes[node] will return all attributes of node
    
nx.draw_networkx_labels(G, pos, labels=node_labels)

plt.show()
```


### node-attributed graphs(GraKeL)


```python
# Creates a list of two node-attributed graphs
G1 = nx.Graph()
G1.add_nodes_from([0,1,2])
G1.add_edges_from([(0,1), (1,2)])
nx.set_node_attributes(G1, {0:np.array([1.1, 0.8]), 
	1:np.array([0.2, -0.3]), 2:np.array([0.9, 1.0])}, 'attributes')

G2 = nx.Graph()
G2.add_nodes_from([0,1,2])
G2.add_edges_from([(0,1), (0,2), (1,2)])
nx.set_node_attributes(G2, {0:np.array([1.8, 0.5]), 
	1:np.array([-0.1, 0.2]), 2:np.array([2.3, 1.2])}, 'attributes')

G_nx = [G1, G2]

# 將 NetworkX圖列表 轉換為 GraKeL圖列表
G_2 = graph_from_networkx(G_nx, node_labels_tag='attributes')


print("2 - Node-attributed graphs transformed")
```

```python
# node-attributed graphs 格式
list(G_2)
```

```python
G=G2
# node-attributed graphs 繪圖
plt.figure(dpi=80,figsize=(6, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos)

#Generate node_labels manually
node_labels = {}
for node in G.nodes:
    node_labels[node] = G.nodes[node] #G.nodes[node] will return all attributes of node
    
nx.draw_networkx_labels(G, pos, labels=node_labels)

plt.show()
```


### edge-label graphs(GraKeL)


```python
# Creates a list of two node-attributed graphs
G1 = nx.Graph()
G1.add_nodes_from([0,1,2])
G1.add_edges_from([(0,1), (1,2)])
nx.set_edge_attributes(G1, 
                       {(0, 1):'a',
                        (1, 2):'b'},'labels')

G2 = nx.Graph()
G2.add_nodes_from([0,1,2])
G2.add_edges_from([(0,1), (0,2), (1,2)])
nx.set_edge_attributes(G2, 
                       {(0,1):'a',
                        (0,2):'b',
                        (1,2):'c'}, 'labels')

G_nx = [G1, G2]

# 將 NetworkX圖列表 轉換為 GraKeL圖列表
G_3= graph_from_networkx(G_nx, edge_labels_tag='labels')


print("3 - edge-label graphs transformed")
```

```python
list(G_3)
```

```python
G=G2
#draw graph with labels
plt.figure(dpi=100)
pos = nx.spring_layout(G)
nx.draw(G, pos)


edge_labels = nx.get_edge_attributes(G, 'labels')
nx.draw_networkx_edge_labels(G, pos, labels=edge_labels)
plt.show()
```


### edge-attributed graphs(GraKeL)


```python
# Creates a list of two node-attributed graphs
G1 = nx.Graph()
G1.add_nodes_from([0,1,2])
G1.add_edges_from([(0,1), (1,2)])
nx.set_edge_attributes(G1, 
                       {(0, 1):np.array([1.1, 0.8]),
                        (1, 2):np.array([0.2, -0.3])},'attributes')

G2 = nx.Graph()
G2.add_nodes_from([0,1,2])
G2.add_edges_from([(0,1), (0,2), (1,2)])
nx.set_edge_attributes(G2, 
                       {(0,1):np.array([1.8, 0.5]),
                        (0,2):np.array([-0.1, 0.2]),
                        (1,2):np.array([2.3, 1.2])}, 'attributes')

G_nx = [G1, G2]

# 將 NetworkX圖列表 轉換為 GraKeL圖列表
G_4 = graph_from_networkx(G_nx, edge_labels_tag='attributes')


print("4 - edge-attributed graphs transformed")
```


### *edge-attributed & node -label graphs(GraKeL)


```python
# Creates a list of two node-attributed graphs
G1 = nx.Graph()
G1.add_nodes_from([0,1,2])
G1.add_edges_from([(0,1), (1,2)])
nx.set_edge_attributes(G1, 
                       {(0, 1):np.array(['ADVMOD', 'Forword']),
                        (1, 2):np.array(['DOBJ', 'Backword'])},'attributes')
nx.set_node_attributes(G1, {0:'ADV', 1:'VERB', 2:'NOUN'}, 'label')
G2 = nx.Graph()
G2.add_nodes_from([0,1,2])
G2.add_edges_from([(0,1), (0,2), (1,2)])
nx.set_edge_attributes(G2, 
                       {(0,1):np.array([1.8, 0.5]),
                        (0,2):np.array([-0.1, 0.2]),
                        (1,2):np.array([2.3, 1.2])}, 'attributes')
nx.set_node_attributes(G2, {0:'a', 1:'b', 2:'c'}, 'label')


G_nx = [G1, G2]

# 將 NetworkX圖列表 轉換為 GraKeL圖列表
G_5 = graph_from_networkx(G_nx, edge_labels_tag='attributes', node_labels_tag='label')


print("5 - edge-attributed & node -label graphs transformed")
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab

G = nx.DiGraph()

G.add_edges_from([('A', 'B'),('C','D'),('G','D')], weight=1)
G.add_edges_from([('D','A'),('D','E'),('B','D'),('D','E')], weight=2)
G.add_edges_from([('B','C'),('E','F')], weight=3)
G.add_edges_from([('C','F')], weight=4)


val_map = {'A': 1.0,
                   'D': 0.5714285714285714,
                              'H': 0.0}

values = [val_map.get(node, 0.45) for node in G.nodes()]
edge_labels=dict([((u,v,),d['weight'])
                 for u,v,d in G.edges(data=True)])
red_edges = [('C','D'),('D','A')]
edge_colors = ['black' if not edge in red_edges else 'red' for edge in G.edges()]

pos=nx.spring_layout(G)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
nx.draw(G,pos, node_color = values, node_size=1500,edge_color=edge_colors,edge_cmap=plt.cm.Reds)
pylab.show()
```
