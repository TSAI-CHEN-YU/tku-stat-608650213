## 範例-兩實體的最短路徑
[exampe sentence] : however, TNF-alpha inhibited myogenin mRNA and protein expression.

```python
doc = nlp('however, TNF-alpha inhibited myogenin mRNA and protein expression.')
for sent in doc.sents:
   print("[position]: ", sent.start, sent.end)
   print("[sentence]: ", sent)
   displacy.render(sent, jupyter=True,style='ent')
```

```python
#依賴解析圖
displacy.render(doc, style='dep', jupyter=True, options={"compact": True})
```

```python
# Label依賴關係(G)
import networkx as nx
import pylab 
import numpy as np

for token in doc:
    print((token.head.text, token.text, token.dep_,token.pos_))
edges = []
for token in doc:
    for child in token.children:
        edges.append(('{0}'.format(token.lower_),
                      '{0}'.format(child.lower_)))

G = nx.Graph(edges)
```

```python
# 最短依賴路徑
entity1 = 'TNF-alpha'.lower()
entity2 = 'myogenin'.lower()

print('原始句子："{}"'.format(doc))

distance=nx.shortest_path_length(G, source=entity1, target=entity2)
print('實體{0}到{1}的距離：'.format(entity1,entity2), distance)

path=nx.shortest_path(G,entity1,entity2)
print('實體{0}到{1}的路徑：'.format(entity1,entity2), path)
```

## 應用-取包含兩個實體的句子

```python
## 包含兩個實體的句子
PMID_sentence = [] #PMID
loc_sentence = [] #PMID
two_ent_sentence = [] #所有被分類的句子
DNA_in_sentence = [] #裡面的DNA實體
for n in range(0,len(df)): #所有摘要(range(0,100)，range(0,len(two_entdf_sentence)))
    doc = nlp(df['Abs'][n]) #第幾篇摘要
    for sent in doc.sents: #摘要的第幾個句子
        doc_sent = doc[sent.start:sent.end] #句子的位置
        if doc_sent.ents != []: # 取所有包含兩個實體的句子
            n_in_sent = []
            all_n = [j.label_ for j in doc_sent.ents]
            for X in doc_sent.ents: # 迴圈遍歷
                if X.label_ == 'DNA': #只取DNA實體
                    all_n.remove(X.label_)
                    n_in_sent.append(X) #句子包含哪些實體(DNA)
                    if len(n_in_sent) >= 2: #若句子包含2個以上的DNA實體
                        if len(n_in_sent) == 2:
                            two_ent_sentence.append(doc_sent) #將這些句子儲存
                            PMID_sentence.append(df['PMID'][n]) #將這些句子儲存
                            loc_sentence.append([sent.start,sent.end])
                            print('第 {} 篇摘要'.format(n+1) , '({})'.format(df['PMID'][n])) 
                            print("<two_ent_sentence> : {}".format(doc_sent))

                    if 'DNA' not in all_n and len(n_in_sent) >= 2:
                        print('包含 {} 個DNA實體:'.format(len(n_in_sent)),n_in_sent)
                        print('-------------------------------------------------------')
                        DNA_in_sentence.append(n_in_sent)
                        
```

```python
# PMID、DNA、Sentence整合成DF
pre_two_path = pd.DataFrame({"PMID": PMID_sentence, "LOCATION": loc_sentence, "SENTENCE": two_ent_sentence, "DNA": DNA_in_sentence  }) # 指定欄標籤名稱 
pre_two_path
```

```python
# 每個包含兩個(以上)實體的句子
print(pre_two_path["SENTENCE"])
```

```python
# 每個句子摘要中的實體
print(pre_two_path["DNA"])
```

## 應用-句子的實體間最短依賴路徑(SDP)子圖

### 單一句子示範
https://pubmed.ncbi.nlm.nih.gov/30055620/

```python
pre_two_path["PMID"][pre_two_path["PMID"]==30055620]
```

```python
# 句子來源位置
n=3588 #(3587)PMID : 30055620
print('\033[1m'+'PMID :','\033[0m', pre_two_path['PMID'][n], pre_two_path['LOCATION'][n]) #PMID
doc = pre_two_path['SENTENCE'][n][:]
DNA = pre_two_path['DNA'][n][:]
print('\033[1m'+'Sent :','\033[0m',doc) #Sentence
print('\033[1m'+'DNA Entities :','\033[0m',DNA,"\n") #DNA Entities

# SENTENCE依賴項標籤 in networkx圖
print('\033[1m'+'Dep Label(father,token,dep,pos) :','\033[0m') #DNA Entities
for token in doc:
    print((token.head.text, token.text, token.dep_,token.pos_))

# 句子依賴關係->networkx圖(G)
edges = []
for token in doc:
    for child in token.children:
        edges.append(('{0}'.format(token.lower_),
                      '{0}'.format(child.lower_)))

G = nx.Graph(edges) 
```

```python
# 分割有空格的實體
print(DNA)
for check in DNA:
    if ' ' in str(check):
        DNA.remove(check)
        print('需分割實體:',check)
        Sp = str(check).split(' ')
        for sp in Sp:
            DNA.append(sp)
print(DNA)
```

```python
from itertools import combinations
# 遍歷句子中實體對的所有組合
print('原始句子："{}"'.format(doc), '\n')
two_ent_comb = list(combinations(DNA, 2 ))
for D in range(len(two_ent_comb)):
    print('組合 :',two_ent_comb[D])
    entity1 = str(two_ent_comb[D][0]).lower()
    entity2 = str(two_ent_comb[D][1]).lower()

    distance=nx.shortest_path_length(G, source=entity1, target=entity2)
    print('實體{0}到{1}的距離：'.format(entity1,entity2), distance)

    path=nx.shortest_path(G, source=entity1, target=entity2)
    print('實體{0}到{1}的路徑：'.format(entity1,entity2), path, '\n')
    
    
```
