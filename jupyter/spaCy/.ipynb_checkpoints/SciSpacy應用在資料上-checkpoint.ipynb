# 結構化處理-獲取摘要文本

```python
#獲取單篇摘要
abs_start = df.find('AB  - ') + 5
print(abs_start) #獲取摘要index,開始

abs_end = df.find('FAU - ') - 1
print(abs_end) #獲取摘要index,結束

print(df[abs_start:abs_end]) #[Arterial Hypertension and Methylenetetrahydrofolate Reductase C677T Gene Polymorphism]
```

```python
# https://blog.csdn.net/FOKJECKT/article/details/95524370?utm_medium=distribute.pc_relevant.none-task-blog-BlogCommendFromMachineLearnPai2-3.channel_param&depth_1-utm_source=distribute.pc_relevant.none-task-blog-BlogCommendFromMachineLearnPai2-3.channel_param
# 定義函數-查詢多個字串
def indexMany(s,str): #str是要查詢的字符
    length = len(s) #獲取該字符串的長度
    str1 = s #拷貝字符串
    list = []
    sum = 0 #用來計算每次截取完字符串的總長度
    try:
        while str1.index(str)!=-1: #當字符串中沒有該字符則跳出
            n = str1.index(str) #查詢查找字符的索引
            str2 = str1[0:n + 1] #截取的前半部分
            str1 = str1[n + 1:length] #截取的後半部分
            sum = sum + len(str2) #計算每次截取完字符串的總長度
            list.append(sum - 1) #把所有索引添加到列表中
            length=length-len(str2) #截取後半部分的長度
    except ValueError:
        return list
    return list
```

```python
#獲取多篇摘要
abs_start = indexMany(df,"AB  - ")
abs_end  = indexMany(df,'FAU - ')
Title_star = indexMany(df,'TI  - ')
Title_end  = indexMany(df,'.\n')
PMID_star = indexMany(df,'PMID- ')
PMID_end  = indexMany(df,'OWN - ')
#save
Title = []
PMID = []
Abs = []
for j in range(0,len(abs_start)):
  # 摘要開頭index
  start = abs_start[j] + 5
  # 摘要結尾index
  c = [i for i in range(len(abs_end)) if abs_end[i]>abs_start[j]]
  end = abs_end[c[0]:c[-1]][0]
  # 標題開頭index
  x = Title_star[j] + 5
  # 標題結尾index
  cc = [i for i in range(len(Title_end)) if Title_end[i]>Title_star[j]]
  y = Title_end[cc[0]:cc[-1]][0]
  # PMID開頭index
  m = PMID_star[j] + 5
  # PMID結尾index
  n = PMID_end[j] - 1
  Title.append(' '.join(df[x:y].replace('\n', '').replace('\r', '').split()))
  PMID.append(df[m:n])
  Abs.append(' '.join(df[start:end].replace('\n', '').replace('\r', '').split()))

  print("Title:",' '.join(df[x:y].replace('\n', '').replace('\r', '').split()))
  print("PMID:",df[m:n]) 
  print("Abs:",' '.join(df[start:end].replace('\n', '').replace('\r', '').split()))
  print('\n')
  print('\n')
```

```python
df = pd.DataFrame([PMID,Title,Abs]).transpose()
df.columns = ["PMID", "Title", "Abs"]
df
```

# *全文章NER結果儲存 (waiting)
**<PubMed_Migrainege-gene_Entity_all.csv>**  ；
**<PubMed_diabetesge-gene_Entity_all.csv>**

```python
nlp_jn = en_ner_jnlpba_md.load()
def add_jn(AbsList, PMIDList):
    i = 0
    table= {"PMID":[], "Entity":[], "Class":[]}
    for doc in nlp_jn.pipe(AbsList):
        PMID = PMIDList[i]
        for x in doc.ents:
          table["PMID"].append(PMID)
          table["Entity"].append(x.text)
          table["Class"].append(x.label_)
        i +=1
    return table
```

```python
#Sort out blank abstracts
import pandas as pd
DF_PMID = pd.Series(PMID)
DF_Abs = pd.Series(Abs)

#Create lists
PMIDList = DF_PMID.tolist()
abstractList = DF_Abs.tolist()

#Add all entity value pairs to table
table = add_jn(abstractList, PMIDList)

#Turn table into an exportable CSV file (returns normalized file of entity/value pairs)
trans_df = pd.DataFrame(table)
trans_df.to_csv ("PubMed_diabetesge-gene_Entity_all.csv", index=False)
```

```python
# Simple usage(範例)
from stanfordcorenlp import StanfordCoreNLP

nlp = StanfordCoreNLP(r'stanford-corenlp-4.1.0', memory='8g')

sentence = 'Guangdong University of Foreign Studies is located in Guangzhou.'
print('Tokenize:', nlp.word_tokenize(sentence))
print('Part of Speech:', nlp.pos_tag(sentence))
print('Named Entities:', nlp.ner(sentence))
print('Constituency Parsing:', nlp.parse(sentence))
print('Dependency Parsing:', nlp.dependency_parse(sentence))

nlp.close() #記得關閉***************
```
