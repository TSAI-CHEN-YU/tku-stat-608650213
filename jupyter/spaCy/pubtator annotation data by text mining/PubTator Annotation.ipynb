## web crawler(網路爬蟲) in PubTator API


```python
import requests
from bs4 import BeautifulSoup## 獲取PubTator註釋
```

```python
# Step1.取得網頁連線
url = 'https://www.ncbi.nlm.nih.gov/research/pubtator-api/publications/export/pubtator?pmids=30055620'
r = requests.get(url)
if r.status_code==200:
  print("正常")
elif r.status_code==400:
  print("不正常")
```

```python
# Step2.使用bs4回傳網頁(html)內容
resp = BeautifulSoup(r.text,"lxml")
print(resp.prettify())
```

```python
# Step3.取得標題所在的標籤內容
title = resp.find('p')
print(title.text)
```

## 提取實體信息 in PubTator API
https://www.ncbi.nlm.nih.gov/CBBresearch/Lu/Demo/PubTator/tutorial/index.html

```python
Abs = title.text[:]
print(Abs)
```

```python
Abs
```

```python
abs_aplit = Abs.split('\n') #
for c in range(0,len(abs_aplit)):
    print('\033[1m',c,'\033[0m',  abs_aplit[c],  '\n')
```

```python
# PMID\START OFFSET\END OFFSET\text MENTION\mention TYPE (e.g. Gene)\database IDENTIFIER. 
import pandas as pd
NER_DF = pd.DataFrame(columns=['PMID','START','END','ENTITY','TYPE','database_IDENTIFIER.'])
for c in range(2,14):
    NER_col = abs_aplit[c].split('\t')
    NER_DF.loc[len(NER_DF)] = NER_col

NER_DF
```

```python
# 所有摘要的PMID(https://pubmed.ncbi.nlm.nih.gov/)
f = open('pmid-diabetes_gene-2001-2021.txt', 'r')
PMID = f.read().splitlines()  #分行讀取
PMID
```

```python
import time
import random
for p in range(0,len(PMID)):
    
    ## Step1.取得網頁連線
    url = 'https://www.ncbi.nlm.nih.gov/research/pubtator-api/publications/export/pubtator?pmids={}'.format(PMID[p])
    slp = random.random()**random.random()*random.random()
    print(p,url,round(slp,3))
    r = requests.get(url)
    # 連線是否正常
    if r.status_code!=200:
        print(p,PMID[p],"不正常")
        continue
    # 休息才能走更遠的路
    time.sleep(slp)
    if p==2000 or p==6700 or p==9000:
        time.sleep(60)
    if p==3800 or p==5000  or p==8000:
        time.sleep(120)
    
    
    # Step2.使用bs4回傳網頁(html)內容
    resp = BeautifulSoup(r.text,"lxml")


    # Step3.取得標題所在的標籤內容
    title = resp.find('p')
    Abs = title.text[:]
    
    # step4.儲存所有PubTator結果
    globals()[str(PMID[p])+"_pubtator"] = Abs
```
